````python
import advertools as adv
import pandas as pd
import plotly.express as px
pd.options.display.max_columns = None

from IPython.display import display_markdown

def md(text):
    return display_markdown(text, raw=True)
````

## Crawl the website

````python
adv.crawl(
    url_list='https://www.nasa.gov/',
    output_file='nasa_crawl.jl',
    follow_links=True,
    custom_settings={
        'CLOSESPIDER_PAGECOUNT': 1000,
        'LOG_FILE': 'nasa_crawl.log',
        'JOBDIR': 'nasa_crawl',
        'AUTOTHROTTLE_ENABLED': True,
        'AUTOTHROTTLE_TARGET_CONCURRENCY': 6
    })
````

````python
crawldf = pd.read_parquet('nasa_crawl.parquet')
crawldf.head(3)
````

````python
md(f'## Rows, columns: {crawldf.shape}')
````

## Summarize images with `advertools.crawlytics.images`

```python
adv.crawlytics.images(crawldf)
```

````python
# img_df.to_parquet('nasa_img_df.parquet', index=False, version='2.6')
# img_df = pd.read_parquet('nasa_img_df.parquet')
````

````python
img_df = adv.crawlytics.images(crawldf)
img_df.head()
````

````python
md(f'## Number of images, attributes: {img_df.shape}')
````

## Number of unique images (`src` attribute)

````python
md(f'#### {img_df['img_src'].nunique():,}')
````

## Distribution of image attribute usage across the site

#### percentage

````python
(img_df
 .replace('', pd.NA)
 .notna()
 .mean()
 .sort_values(ascending=False)
 .reset_index()
 .rename(columns={'index': 'attribute', 0: '%'})
 .style
 .format({'%': '{:.1%}'})
 .bar(color='steelblue', subset=['%'])
 # .set_caption('<h4>img atribute usage - percentage</h4')
)
````

#### count

````python
(img_df
 .replace('', pd.NA)
 .notna()
 .sum()
 .sort_values(ascending=False)
 .reset_index()
 .rename(columns={'index': 'attribute', 0: 'count'})
 .style
 .format({'count': '{:,}'})
 .bar(color='steelblue', subset=['count'])
 # .set_caption('<h4>img atribute usage - count</h4')
)
````

## De-duplicated

````python
(img_df
 .drop_duplicates(subset=['img_src'])
 .replace('', pd.NA)
 .notna()
 .mean()
 .sort_values(ascending=False)
 .reset_index()
 .rename(columns={'index': 'attribute', 0: '%'})
 .style
 .format({'%': '{:.1%}'})
 .bar(color='steelblue', subset=['%'])
 # .set_caption('<h4>img atribute usage – deduplicated</h4')
)
````

````python
import re
def img_dim(measure):
    if pd.isna(measure):
        return None
    if measure.isdigit():
        return float(measure)
    if isinstance(measure, (int, float)):
        return measure
    if 'px' in measure:
        number = re.findall(r'\d+', measure)[0]
        return number
    if 'rem' in measure:
        number = re.findall(r'\d+', measure)[0]
        return number * 16
    if measure == '':
        return None

width = [img_dim(width) for width in img_df['img_width']]
height = [img_dim(height) for height in img_df['img_height']]
````

## Image alt attribute ngrams

````python
(adv.word_frequency(
    img_df['img_alt']
    .fillna(''))
 .head(20)
 .style
.format({'abs_freq': '{:,}'})
.bar(subset=['abs_freq'], color='steelblue'))
````

````python
(img_df
 [img_df['img_alt']
 .fillna('')
 .str.contains('Orange', case=False)]
 [['url', 'img_alt']]
 ['img_alt']
 .value_counts()
 .reset_index())
````

````python
rm_words = [
    'of the', 'on the', 'to the', 'with a', 'is a', 'to a', 'in the', 'for the',
    'on a', 'inside the', 'a black', 'the left', '']
tri_grams =adv.word_frequency(
    img_df['img_alt'].fillna(''),
    phrase_len=3,
    rm_words=rm_words)
````
