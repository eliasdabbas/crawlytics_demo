```python
def compare(df1, df2, column, keep_equal=False):
    """Compare common URLs in two crawl DataFrames with respect to `column`.

    Parameters
    ----------
    df1 : pandas.DataFrame
      The DataFrame of the first crawl
    df2 : pandas.DataFrame
      The DataFrame of the second crawl
    column : str
      The name of the column according to which you want to compare
    keep_equal : bool, default False
      Whether or not to keep unchanged values in the result

    Returns
    -------
      comparison_df : pandasDataFrame

    Examples
    --------

    >>> import advertools as adv
    >>> import pandas as pd
    >>> df1 = pd.read_json('output_file1.jl', lines=True)
    >>> df2 = pd.read_json('output_file2.jl', lines=True)
    >>> adv.crawlytics.compare(df1, df1, 'size')

====  ==========================  ========  ========  ======  ===========
  ..  url                           size_x    size_y    diff    diff_perc
====  ==========================  ========  ========  ======  ===========
   0  https://example.com/page_1    299218    317541   18323    0.0612363
   1  https://example.com/page_2    214891    208886   -6005   -0.0279444
   2  https://example.com/page_3    257442    251437   -6005   -0.0233256
   3  https://example.com/page_4    230403    224398   -6005   -0.026063
   4  https://example.com/page_5    222242    216237   -6005   -0.0270201
====  ==========================  ========  ========  ======  ===========
    """
    compare_df = (pd.merge(
        df1[['url', column]],
        df2[['url', column]],
        left_on='url',
        right_on='url')
     .assign(changed=lambda df: df[f'{column}_x'].ne(df[f'{column}_y'])))
    if ('int' in str(df1[column].dtype).lower()) or ('float' in str(df1[column].dtype).lower()):
        compare_df['diff'] = compare_df[f'{column}_y'].sub(compare_df[f'{column}_x'])
        compare_df['diff_perc'] = compare_df['diff'].div(compare_df[f'{column}_x'])
    compare_df = compare_df.dropna(thresh=compare_df.shape[1])
    if keep_equal:
        return compare_df
    else:
        return compare_df[compare_df['changed']].drop('changed', axis=1)
```

```python
dflist = {}
errors = []
for col in df1.columns[1:]:
    try:
        df = compare(df1, df2, col)
        dflist[col] = df
    except Exception as e:
        errors.append((col, str(e)))
        continue
```

```python
for column, diff_df in dflist.items():
    if not dflist[column].empty:
        print()
        display(diff_df)
        print()
```
